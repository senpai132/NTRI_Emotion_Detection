```python
import pandas as pd
```

```python
emotion_df = pd.read_csv('datasets/emotion-dataset.csv')
emotion_df
```

```python
tweet_df = pd.read_csv('datasets/tweet_emotions.csv')
tweet_df
```

```python
emotion_df['Emotion'].unique()
```

```python
tweet_df['sentiment'].unique()
```

```python
extracted_tweet_df = tweet_df[['sentiment', 'content']]
extracted_tweet_df
```

```python
extracted_tweet_df_renamed = extracted_tweet_df.rename(columns = {'sentiment':'Emotion', 'content':'Text'})
extracted_tweet_df_renamed
```

```python
merged_df = pd.concat([emotion_df, extracted_tweet_df_renamed])
merged_df
```

```python
merged_df = merged_df.loc[merged_df['Emotion']!='neutral']
merged_df
```

```python
merged_df.to_csv(path_or_buf = 'datasets/merged.csv', columns = ['Emotion', 'Text'],
                 index = False)
```

```python
df = pd.read_csv('datasets/merged.csv')
df
```

```python
df['Emotion'].unique()
```

```python
df['Emotion'].value_counts()
```

```python
df_binary = df.replace(['joy', 'surprise', 'enthusiasm', 'love', 'fun', 'happiness', 'relief'],'positive')
df_binary = df_binary.replace(['sadness', 'fear', 'anger', 'shame', 'disgust', 'empty', 'worry', 'hate', 'boredom'], 'negative')
df_binary
```

```python
df_binary['Emotion'].value_counts()
```

```python
df_binary.to_csv(path_or_buf = 'datasets/binary-sentiment.csv', columns = ['Emotion', 'Text'],
                 index = False)
```
